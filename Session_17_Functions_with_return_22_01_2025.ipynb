```python
# wap ask the user enter the number
# print it is even or odd
# a) Function without argument with default and with argument

def even_odd():   #without argument
    n=eval(input("enter number:"))
    if n%2==0:
        print("Even")
    else:
        print("odd")

even_odd()       
```

```python
def even_odd1(num):    # with argument
    if num%2==0:
        print("Even")
    else:
        print("odd")

even_odd1(105)
```

```python
def even_odd2(num1=104):    # default argument
    if num1%2==0:
        print("Even")
    else:
        print("odd")

even_odd2()
```

```python
# Q1) factorial recursion 5! = 5*4*3*2*1

def factorial(n):
    if n==0:
        return(1)
    else:
        return(n*factorial(n-1))
        
a=factorial(5)
print(a)

```

```python
# Q2) Calculator

def add(a,b):
    return(a+b)
def sub(a,b):
    return(a-b)
def mul(a,b):
    return(a*b)
def div(a,b):
    return(a/b)
print("enter 1 for add")
print("enter 2 for sub")
print("enter 3 for mul")
print("enter 4 for div")

def main():
    option=eval(input('enter between 1 and 4:'))
    if option==1:
        a=eval(input("enter a:"))
        b=eval(input("enter b:"))
        add(a,b)
        return(a+b)
    elif option==2:
        a=eval(input("enter a:"))
        b=eval(input("enter b:"))
        sub(a,b)
        return(a-b)
    elif option==3:
        a=eval(input("enter a:"))
        b=eval(input("enter b:"))
        mul(a,b)
        return(a*b)
    elif option==4:
        a=eval(input("enter a:"))
        b=eval(input("enter b:"))
        div(a,b)
        return(a/b)
    else:
        return("Incorrect Number") 
main()
        
        
        
```

```python
# Q2) Calculator

def add():
    a=eval(input("enter a value:"))
    b=eval(input("enter b value:"))
    return(a+b)
def sub():
    a=eval(input("enter a value:"))
    b=eval(input("enter b value:"))
    return(a-b)
def mul():
    a=eval(input("enter a value:"))
    b=eval(input("enter b value:"))
    return(a*b)
def div():
    a=eval(input("enter a value:"))
    b=eval(input("enter b value:"))
    return(a/b)
print("enter 1 for add,enter 2 for sub,enter 3 for mul, enter 4 for div")

def main():
    option=eval(input('enter between 1 and 4:'))
    if option==1:
        result=add()
        print(result)
    elif option==2:
        result=sub()
        print(result)
    elif option==3:
        result=mul()
        print(result)
    elif option==4:
        result=div()
        print(result)
    else:
        return("Incorrect Number") 
main()
        
        
        
```
